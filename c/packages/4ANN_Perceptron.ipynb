```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.array([ [0,0] , [0,1] , [1,0] , [1,1] ])
Y = np.array([ -1 , -1 , -1 , 1])
W = np.zeros(2)
print(X , Y , W)
```

```python
for _ in range(6):
    for i in range(4):
        output = np.dot( X[i] , W)
        if output >= 0:
            output = 1
        else:
            output = 0
        error = Y[i] - output
        W +=  0.1 * error * np.array(X[i])
        print(W)
```
